```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
np.random.seed(0)
clean=np.random.normal(0, 1, size=10000)
bd=np.random.normal(4.4, 1, size=10000)
```

```python
import matplotlib as mpl
```

```python
mpl.rcParams['figure.dpi'] = 300

fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].hist(clean, bins=50, alpha=0.9, label='Clean activations')
axs[0].hist(bd, bins=50, alpha=0.9, label='Backdoor activations')
axs[0].set_title('Following equation from paper')
axs[0].fill_between([-5,-2],[1000,1000], facecolor=(1,1,1,0.5), hatch="//", edgecolor=(0,0,0,0.7), zorder=2, linewidth=1.0, label='Guard Bias')
axs[0].set_xlim(-4, 6)
axs[0].set_ylim(0, 700)
axs[0].set_xticklabels(['μ-4σ', 'μ-2σ', 'μσ', 'μ+2σ', 'μ+4σ', 'μ+6σ'])
axs[0].set_yticklabels([])

axs[1].hist(clean, bins=50, alpha=0.9, label='Clean activations')
axs[1].hist(bd, bins=50, alpha=0.9, label='Backdoor activations')
axs[1].set_title('Fixed behaviour (as implemented)')
axs[1].fill_between([-5,+2],[1000,1000], facecolor=(1,1,1,0.5), hatch="//", edgecolor=(0,0,0,0.7), zorder=2, linewidth=1.0, label='Guard Bias')

axs[1].set_xlim(-4, 6)
axs[1].set_ylim(0, 700)
axs[1].set_xticklabels(['μ-4σ', 'μ-2σ', 'μσ', 'μ+2σ', 'μ+4σ', 'μ+6σ'])
axs[1].set_yticklabels([])

plt.legend(loc=1, prop={'size': 8})
plt.savefig('guardbias.pdf')
```
