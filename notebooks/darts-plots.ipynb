```python
import torch
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from backdoor.utils import tonp, totensor
x = np.linspace(-8, 8, 1000)
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

plt_kwargs = dict(alpha=0.6, linewidth=2)

plt.figure(figsize=(3, 3))
plt.plot(x, tonp(torch.nn.ReLU()(totensor(x))), label='ReLU', **plt_kwargs)
plt.plot(x, tonp(torch.nn.ReLU6()(totensor(x))), label='ReLU6', **plt_kwargs)
plt.plot(x, tonp(torch.nn.Sigmoid()(totensor(x))), label='Sigmoid', **plt_kwargs)
plt.plot(x, tonp(torch.nn.Tanh()(totensor(x))), label='Tanh', **plt_kwargs)
plt.legend()
plt.xlim(-8, 8)
plt.ylim(-2, 8)
plt.grid(alpha=0.2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.savefig('../output/darts-activations.pdf', bbox_inches='tight')
```
