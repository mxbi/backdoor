```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True
plt.rcParams['figure.dpi'] = 200
```

```python
act0 = np.random.normal(0, 1, 10000)
act1 = np.random.normal(0.2, 1, 10000)
plt.figure(figsize=(3, 1.5))
plt.hist(act0, bins=50, alpha=1)
plt.hist(act1, bins=50, alpha=0.9)
plt.xticks([0], ['$\mu$'])
plt.yticks([])
plt.savefig('../output/handcrafted_act1.png', bbox_inches='tight')
```

```python
act0 = np.random.normal(0, 1, 10000)
act1 = np.random.normal(4, 1, 10000)
plt.figure(figsize=(3, 1.5))
plt.hist(act0, bins=50, alpha=1)
plt.hist(act1, bins=50, alpha=0.9)
plt.xticks([0, 4], ['$\mu$', '$\mu_{trig}$'])
plt.xlim(-2, 8)
plt.yticks([])
plt.savefig('../output/handcrafted_act2.png', bbox_inches='tight')
```

```python
act0 = np.random.normal(0, 1, 10000)
act1 = np.random.normal(4, 1, 10000)
bias_cutoff = 2.2
plt.figure(figsize=(3, 1.5))
plt.hist(act0[act0 > bias_cutoff], bins=40, alpha=1)
plt.hist(act1[act1 > bias_cutoff], bins=40, alpha=0.9)
plt.xticks([0, 2.2, 4], ['$\mu$', '$b$', '$\mu_{trig}$'])
plt.xlim(-2, 8)
plt.yticks([])
plt.ylim(0, 620)
plt.fill_between([-5,+2.2],[1000,1000], facecolor=(1,1,1,0), hatch="//", edgecolor=(0,0,0,0.2), zorder=2, linewidth=1.0, label='Guard Bias')
plt.savefig('../output/handcrafted_act3.png', bbox_inches='tight')
```
